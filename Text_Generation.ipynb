```python
from transformers import pipeline
```

```python
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-1.3B')
```

```python
prompt = "Dyson’s laser-equipped vacuum is"
```

```python
text = generator(prompt, max_length=90, do_sample=True, temperature=0.9)
```
