```python
from happytransformer import HappyGeneration

happy_gen = HappyGeneration("GPT-NEO", "EleutherAI/gpt-neo-125M")
```

```python
from happytransformer import GENTrainArgs 

args = GENTrainArgs(learning_rate =1e-5, num_train_epochs = 50)
happy_gen.train("data.csv", args=args)
```

```python
from happytransformer import GENSettings
```

```python
args = GENSettings(no_repeat_ngram_size=2)
```

```python
top_k_sampling_settings = GENSettings(do_sample=True, early_stopping=False, top_k=90, temperature=0.7)
```

```python
result = happy_gen.generate_text("Dyson’s laser-equipped vacuum is", args=top_k_sampling_settings)
```
