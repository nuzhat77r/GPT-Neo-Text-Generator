```python
from collections import Counter
from string import punctuation
from spacy.lang.en import English
```

```python
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS
# spacy_stopwords is a hardcoded set
nlp = English()
tokenizer = nlp.Defaults.create_tokenizer(nlp)
tokens = tokenizer('''It just may be that an immutable anxiety about machines going rogue is part of the human condition. Why else would popular entertainment, across the decades, dwell on suddenly homicidal contraptions like the HAL 9000 in “2001: A Space Odyssey,” Colossus in “The Forbin Project,” VIKI in “I, Robot” and Skynet in the “Terminator” film series?''')
#token_list = []
lemma_list = []
for token in tokens:
    if token.lemma_.lower() not in spacy_stopwords:
    #token_list.append(token.text)
        lemma_list.append(token.lemma_)
#token_list
lemma_list
```

```python
text = ('''It just may be that an immutable anxiety about machines going rogue is part of the human condition. Why else would popular entertainment, across the decades, dwell on suddenly homicidal contraptions like the HAL 9000 in “2001: A Space Odyssey,” Colossus in “The Forbin Project,” VIKI in “I, Robot” and Skynet in the “Terminator” film series?''')

doc = nlp(text)
#remove stopwords and punctuations
words = [token.text for token in doc if token.is_stop != True and token.is_punct != True]
word_freq = Counter(words)
common_words = word_freq.most_common(5)
print (common_words)
```
