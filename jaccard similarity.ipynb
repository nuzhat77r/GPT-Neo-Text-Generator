```python
# documents

Generated_text="grant climate scores can be calculated for the target country based on the country-specific population characteristics, and that the GHRs could be converted to the target values using the following formula ([@bib42]):$$\"

Collected_text="The new climate provisions in the infrastructure bill inject billions of dollars into competitive grant programs. These are pots of money that towns, cities and counties can access only by submitting applications, which federal agencies then rank, with funds going to applicants with the highest scores."

#copy
para = Collected_text
output = Generated_text

# Split the documents and create tokens
doc1_tokens=set(Generated_text.lower().split())
doc2_tokens=set(Collected_text.lower().split())

#Print the tokens
print(doc1_tokens ,doc2_tokens)


# Calculate the Jaccard Similarity
jaccard_similarity=  len(doc1_tokens.intersection(doc2_tokens))/len(doc1_tokens.union(doc2_tokens))

# Print the Jaccard Simialrity score
print("Similarity %:",jaccard_similarity*100)

#finding intersected words
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

#remove special characters
for ele in para:
    if ele in punc:
        para = para.replace(ele, "")

for ele in output:
    if ele in punc:
        output = output.replace(ele, "") 

#lowercase
para = para.lower()
output = output.lower()
#intersection
str1_words = set(para.split())
str2_words = set(output.split())
common = str1_words & str2_words #intersection                
print("The following are the intersected words: ")
print(common)
```
