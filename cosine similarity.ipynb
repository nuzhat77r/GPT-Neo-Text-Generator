```python
# documents

Generated_text = 'And it feels bizarre to me that if Flo was a friend, they would have been friends by now. I do believe that Flo is a friend now. I don know if that because Flo has gotten stronger or if her friendship with Lita has really developed or if I was a jerk to her after all these years.'
Collected_text = 'And it feels bizarre to me that if Flo just releases a new privacy policy, it then can share women’s intimate information. But that’s mostly how it works in the United States. Companies can do pretty much whatever they want concerning their users’ data if they first outline their actions in a privacy policy.'

# Split the documents and create tokens
doc1_tokens=set(Generated_text.lower().split())
doc2_tokens=set(Collected_text.lower().split())

#Print the tokens
output= doc1_tokens and doc2_tokens
print("common",doc1_tokens and doc2_tokens)
word_freq = Counter(output)
common_words = word_freq.most_common(3)
print("Word_frequency:", common_words)
```
